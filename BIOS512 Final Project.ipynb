## Analysis of Marriage Dataset

##### The marriage dataset can be accessed using this [link](https://github.com/fivethirtyeight/data/tree/master/marriage). 

```r
library('tidyverse')

marriage = read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/marriage/both_sexes.csv', 
         col_types = cols())
```

```r
marriage1= marriage %>% rename(Obs = X1)
marriage1

marriage2 = marriage1 %>%
    select (year,all_2534, all_3544, all_4554) %>%
    pivot_longer(cols = !c(year), names_to = "Age", values_to = "Rate") %>% 
    mutate (Rate =1-Rate) %>%
    mutate(Age= recode(Age, "all_2534" = "25-34", "all_3544" = "35-44", "all_4554" = "45-54"))

head(marriage2)
```

```r
options(repr.plot.width = 5, repr.plot.height = 3, repr.plot.res = 250)

p = ggplot(data = marriage2, mapping = aes(x = year, y = Rate, color= Age))

p = p + geom_line() + scale_color_brewer(palette = "Set2")

p = p + labs(title="Marriage Rate by Age", 
             x = 'Year', y = 'Fraction of Population Married') + theme_update(plot.title = element_text(hjust = 0.5))
p 
```

We can see from the output of "marriage1," this data set has 17 rows ordered by the year. The values in each cell outside of the "Obs," "year," and "date" columns represent the fraction of the population that has never been married in the specific year. Since I wanted to see the marriage rate, I subtracted the fraction of the population that has never been married from one. After plotting the marriage rates by age group, we can see that the younger group (age 25-34) has the lowest marriage rate. In 1960, about 88% of 25–34 year olds had been married, but that fraction falls drastically about 51% in 2012. While those ages 35-44 and 45-54, have the same percentage of marriage in 1960, which is about 95%, the marriage rate has fallen faster for those ages 35-44, with the fraction married being about 78% and 85% for age group 35-44 and 45-54 respectively.

```r
marriage3 = marriage1 %>%
    select (year, White_2534, Black_2534, Hisp_2534, White_3544, Black_3544, Hisp_3544, 
            White_4554, Black_4554, Hisp_4554) %>%
    pivot_longer(cols = !c(year), names_to = "Race_Age", values_to = "Rate") %>% 
    separate(col=Race_Age, sep = "_", into=c('Race', 'Age'), remove = FALSE) %>% 
    mutate (Rate =1-Rate) %>%
    mutate(Age= recode(Age, "2534" = "25-34", "3544" = "35-44", "4554" = "45-54"))

head (marriage3)
```

```r
options(repr.plot.width = 8, repr.plot.height = 4, repr.plot.res = 250)

p = ggplot(data = marriage3, mapping = aes(x = year, y = Rate, color = Race))

p = p + geom_line() + scale_color_brewer(palette = "Set2", labels = c("Black", "Hispanic", "White"))

p = p + facet_wrap(~Age, nrow = 1)

p = p + labs(title="Marriage Rate by Age and Race", 
             x = 'Year', y = 'Fraction of Population Married') + theme_update(plot.title = element_text(hjust = 0.5))
p 
```

The graph above shows the marriage rate by Race broken down between the age groups.  From this graph, we can see that Black Americans had the lowest marriage rate in all three age groups, followed by Hispanic Americans. In 1960, people in all three races had similar marriage percentage, with it being almost equal in the age group 45-54, but over the years the marriage rate among Black Americans begins to decline faster than the other races. Although the marriage rate of Hispanic Americans was always slightly lower than white Americans in all the age groups, the rate of decline over the years was similar for the two races. For all races the marriage rate is higher for the older age group.

```r
marriage4 = marriage1 %>%
    select (year, poor_2534, mid_2534, rich_2534, poor_3544, mid_3544, rich_3544, 
            poor_4554, mid_4554, rich_4554) %>%
    pivot_longer(cols = !c(year), names_to = "Income_Age", values_to = "Rate") %>% 
    separate(col=Income_Age, sep = "_", into=c('Income', 'Age'), remove = FALSE) %>% 
    mutate (Rate =1-Rate) %>%
    mutate(Age= recode(Age, "2534" = "25-34", "3544" = "35-44", "4554" = "45-54")) %>%
    mutate(Income= recode(Income, "poor" = "Lower Class", "mid" = "Middle Class", "rich" = "Upper Class"))

head (marriage4)
```

```r
options(repr.plot.width = 8, repr.plot.height = 4, repr.plot.res = 250)

p = ggplot(data = marriage4, mapping = aes(x = year, y = Rate, color = Income))

p = p + geom_line() + scale_color_brewer(palette = "Set2")

p = p + facet_wrap(~Age, nrow = 1)

p = p + labs(title="Marriage Rate by Age and Income", 
             x = 'Year', y = 'Fraction of Population Married') + theme_update(plot.title = element_text(hjust = 0.5))
p 
```

The graph above shows the marriage rate by income broken down between the age groups. In 1960 the middle class had the highest marriage percentage in all the age groups, followed by the upper class in age groups 35-44 and 45-54. For adults 25-34, the lower class had the second highest marriage percentage, which changed in 1970, when the marriage rate for the upper class became higher than the lower class. While the marriage rate declined steadily for the middle- and lower-class population, the graph shows both increases and decreases throughout the years for the upper class. From 2000-2005, there seems to be a sudden jump in marriage rate for the upper class in all the age groups. The marriage rate for upper class individuals between the age of 45 and 54 was higher in end (2012) than the beginning (1960), which is not the case for the rest of the income groups.
