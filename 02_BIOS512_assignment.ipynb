Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
p1 = ggplot(penguins, aes(x = body_mass_g))

p1 = p1 + geom_histogram(bins = 50)

p1 = p1 + facet_wrap(~year, nrow = 3) + scale_x_continuous(label = scales::comma) 

p1
```
