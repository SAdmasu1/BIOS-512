Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
library(scales)
options(repr.plot.width = 7, repr.plot.height = 6, repr.plot.res = 250)
p1 = ggplot(penguins, aes(x = body_mass_g))

p1 = p1 + geom_histogram(bins = 40, color="aquamarine4", fill="aquamarine3") + facet_wrap(~year, nrow = 3) 

p1 = p1 + labs(x = "Body Mass in Grams", y = "Number of Penguins",
               title = "Histogram of Penguins Body Mass from Measurements taken from 2007-2009")

p1 = p1 + scale_x_continuous(labels = comma) 

p1 = p1 + theme(strip.background = element_rect(colour = "aquamarine4", fill = "white")) 

p1 = p1 + theme(strip.text.x = element_text(colour = "aquamarine4", face = "bold")) 

p1 = p1 + theme(axis.text.x = element_text(angle = 45, hjust = 1)) 

p1
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

- Yes the male penguins have a higher body mass than the female penguins.

```r
options(repr.plot.width = 7, repr.plot.height = 7, repr.plot.res = 250)
p2 = ggplot(penguins, aes(x = body_mass_g, y=sex))

p2 = p2 + geom_point(alpha = 0.5, position = position_jitter(h = 0.2, w = 0.2), size = 2, color="aquamarine4", fill="aquamarine3")

p2 = p2 + labs(x = "Body Mass in Grams", y = "Sex",
               title = "Penguin Body Mass by Sex")

p2 = p2 + facet_wrap(~species, nrow = 3) 

p2 = p2 + theme(strip.background = element_rect(colour = "aquamarine4", fill = "white")) 

p2 = p2 + theme(strip.text.x = element_text(colour = "aquamarine4", face = "bold")) 

p2 = p2 + theme(plot.title = element_text(hjust = 0.5))

p2
```
