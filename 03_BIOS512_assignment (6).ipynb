In this assignment we'll be using data from the [TidyTuesday Project](https://github.com/rfordatascience/tidytuesday). Specifically, we'll investigate vaccination rates at US schools. A `CSV` file of the data is located at:  

[https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-02-25/measles.csv](https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-02-25/measles.csv)

Get the "raw" data URL and import it into your `R` session using `read_csv`. Remember to capture the data under a variable name of your choosing!

```r
measles = read.csv(
    'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-25/measles.csv', 
    stringsAsFactors = FALSE)

head(measles)
```

The `mmr` column has holds ðŸ¤’ measles, mumps, rubella ðŸ¤’ vaccination rates for students in each respective school. **If the `mmr` value is not available for a school the `mmr` value is set to -1 in this data set.** 

The target `mmr` vaccination rate as [recommended by the CDC](https://www.cdc.gov/mmwr/preview/mmwrhtml/mm5632a3.htm) for schools is 95%. 

**Calculate the fraction of schools per state that have vaccination rates greater than the CDC 95%. Capture the output as a table called `df_vacc_rates`.**

You'll need to do use `filter`, `group_by`, and `summarize`. 

I.e.
1. Filter out schools that don't report `mmr` vaccination rate (keep schools where `mmr >= 0`).
1. Group the data by `state`.
1. Summarize fracion of vaccination rates above 95%.

ðŸ’¡Remember `n()` stands for the number or records in a group. Also, `sum(mmr >= 95, na.rm=TRUE)` will count the number of values above or equal to 95.ðŸ’¡

```r
# create the df_vacc_rates here

df_vacc_rates = measles %>%
    group_by(state) %>%
    select(mmr, state) %>%
    filter(mmr >= 0) %>%
    summarize("Fraction_of_Schools" =sum(mmr >= 95, na.rm=TRUE)/n())

head(df_vacc_rates)
```

Which state (of those that report `mmr`) has the smallest fraction of schools above the CDC target vaccination rate of 95%?  

```r
df_vacc_rates[df_vacc_rates$Fraction_of_Schools==min(df_vacc_rates$Fraction_of_Schools, na.rm=TRUE),]

"Arkansas has the smallest fraction of schools above the CDC target vaccination rate of 95%"
```

**Make a ECDF plot of the school vaccination rates in North Carolina.** Use the `overall` column which reports the "overall" vaccination rate at each school.

â—ï¸Remember, you can calculate the `y` value for ECDF charts using `row_number`, `mutate`, and `arrange`.

```r
ecdf = measles %>%
    filter(state == "North Carolina") %>%
    arrange(overall) %>%
    mutate(rn = row_number(), fraction_of_schools = rn / max(rn))

options(repr.plot.res = 400, repr.plot.height = 4, repr.plot.width = 6)

p = ggplot(ecdf, aes(x = overall, y = fraction_of_schools))

p = p + geom_point(size = 1, alpha=0.2, color="aquamarine4") + geom_step(color="aquamarine4")

p = p + geom_vline(aes(xintercept = 95), color = 'orange')

p = p + geom_text(
    label=stringr::str_wrap("CDC recommended vaccination rate (95%)", width = 5), 
    x = 84, y = 0.75, color = 'orange')

p = p + theme_minimal()

p = p + labs(title="ECDF Plot of School Vaccination Rates in North Carolina", 
             x = 'Vaccination Rate (%)', y = 'Fraction of Schools')


p
```
